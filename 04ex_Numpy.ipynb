2\. Find the outer product of the following two vecotrs

```python

```

Do this in the following ways:

   * Using the function outer in numpy
   * Using a nested for loop or list comprehension
   * Using numpy broadcasting operatoins


````python
u = np.array([1,3,5,7])
v = np.array([2,4,6,8])

# Using the function outer in numpy.

prod1 = np.outer(u,v)
print('The outer product of u y v using the numpy function is:', '\n', prod1, '\n')

# Using a nested for loop or list comprehension.

# l = []
# for i in range(len(u)):
#     row = []
#     for j in range(len(v)):
#         row.append(u[i]*v[j])
#     l.append(row)
# prod21 = np.array(l)
# print(prod21)

prod22 = np.array([u[i]*v[j] for i in range(len(u)) for j in range(len(v))]).reshape(len(u), len(v))
print('The outer product of u y v using a comprehension list is:', '\n', prod22, '\n')

# Using a numpy broadcasting operations.

# prod3 = u[:, None] * v

prod3 = u.reshape(len(u), 1) * v.reshape(1, len(v))
print('The outer product of u y v using numpy broadcasting operations:', '\n', prod3, '\n')
````

7\. Prime numbers sieve: compute the prime numbers in the 0-N (N=99 to start with) range with a sieve (mask).
  * Constract a shape (100,) boolean array, the mask
  * Identify the multiples of each number starting from 2 and set accordingly the corresponding mask element
  * Apply the mask to obtain an array of ordered prime numbers
  * Check the performances (timeit); how does it scale with N?
  * Implement the optimization suggested in the [sieve of Eratosthenes](https://en.wikipedia.org/wiki/Sieve_of_Eratosthenes)

````python
import timeit

# def primes(n):
#     numbers = np.arange(n)
#     mask = np.array(np.zeros(n), bool)
#     for i in range(0,n):
#         if numbers[i] % 2 !=0 and numbers[i] % 3 != 0 and numbers[i] % 5 != 0 and numbers[i] % 7 != 0 and numbers[i] % 11 != 0 and numbers[i] % 13 != 0:
#             mask[i] = True
#     mask[0:2] = False
#     mask[3:9:2] = True
#     mask[11:15:2] = True
#     return numbers[mask]

def primes(n):
    if n < 2: return np.array([])
    nums = np.arange(n)
    limit = int(np.sqrt(n))
    divisors = np.arange(2, limit + 1)
    is_divisible = (nums[:, None] % divisors == 0)
    is_composite = np.any(is_divisible & (nums[:, None] != divisors), axis=1)
    mask = ~is_composite
    mask[0:2] = False
    return nums[mask]

n_values = [100, 500, 1000, 2000, 5000]
results = []

for n in n_values:
    t = timeit.repeat(lambda: primes(n), repeat=3, number=1000)
    avg_single_run = min(t) / 1000
    results.append(avg_single_run)

plt.figure(figsize=(8, 5))
plt.plot(n_values, results, marker='o', linestyle='-', color='teal')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Value of N')
plt.ylabel('Time per single execution (seconds)')
plt.title('Performance Scaling of Vectorized Primes')
plt.grid(True, which="both", ls="-", alpha=0.5)
plt.show()
````

8\. Diffusion using random walk

Consider a simple random walk process: at each step in time, a walker jumps right or left (+1 or -1) with equal probability. The goal is to find the typical distance from the origin of a random walker after a given amount of time. 
To do that, let's simulate many walkers and create a 2D array with each walker as a raw and the actual time evolution as columns

  * Take 1000 walkers and let them walk for 200 steps
  * Use randint to create a 2D array of size walkers x steps with values -1 or 1
  * Build the actual walking distances for each walker (i.e. another 2D array "summing on each raw")
  * Take the square of that 2D array (elementwise)
  * Compute the mean of the squared distances at each step (i.e. the mean along the columns)
  * Plot the average distances (sqrt(distance\*\*2)) as a function of time (step)
  
Did you get what you expected?

````python
import numpy.random as npr
npr.seed(123)

vals = np.random.randint(0, 2, size=(1000, 200)) * 2 - 1
distance = vals.cumsum(axis=1)

square_distance = distance ** 2

mean = square_distance.mean(axis=0)

average_distance = np.sqrt(mean)

t = np.arange(1, 201)
theoretical = np.sqrt(t)

plt.figure(figsize=(8, 5))
plt.plot(np.arange(1, 201), average_distance, label='RMS Distance')
plt.plot(t, theoretical, '--', label='Theoretical ($\sqrt{t}$)', color='red')
plt.title("Diffusion: Typical Distance vs Time")
plt.xlabel("Number of Steps")
plt.ylabel("Typical Distance")
plt.grid(True)
plt.legend()
plt.show()
````

````python
data = np.loadtxt('populations.txt')
year, hares, lynxes, carrots = data.T

plt.figure(figsize=(8, 5))
plt.plot(year, hares, label='hares population over the years', color = 'red')
plt.plot(year, lynxes, label='lynxes population over the years', color = 'blue')
plt.plot(year, carrots, label='carrots population over the years', color = 'yellow')
plt.title("Populations over the years")
plt.xlabel("Year")
plt.ylabel("Population")
plt.legend()
plt.show()
````

````python
year = np.array(year)
hares = np.array(hares)
lynxes = np.array(lynxes)
carrots = np.array(carrots)

variables = [year, hares, lynxes, carrots]
names = ["Year", "Hares", "Lynxes", "Carrots"]

for name, var in zip(names, variables):
    print(f"--- Statistics for {name} ---")
    print(f"Mean:     {np.mean(var):.2f}")
    print(f"Std Dev:  {np.std(var):.2f}")
    print(f"Min:      {np.min(var)}")
    print(f"Max:      {np.max(var)}")
    print(f"Median:   {np.median(var)}")

correlation_matrix = np.corrcoef(data.T)
correlation_matrix
````

````python
maximum = np.max(data[:,1:], axis= 1)
for i in range(len(data)):
    if maximum[i] == hares[i]:
        print(f'The highest poluation in {year[i]} was hares with: {hares[i]} population')
    if maximum[i] == lynxes[i]:
        print(f'The highest poluation in {year[i]} was lynxes with: {lynxes[i]} population')
    if maximum[i] == carrots[i]:
        print(f'The highest poluation in {year[i]} was carrots with: {carrots[i]} population')
````

````python
species_data = data[:, 1:]

# Get the index of the maximum value for each year (axis 1)
# 0 = Hares, 1 = Lynxes, 2 = Carrots
max_indices = np.argmax(species_data, axis=1)

species_names = np.array(['Hares', 'Lynxes', 'Carrots'])
winners = species_names[max_indices]

for yr, name, count in zip(year, winners, maximum):
    print(f"In {int(yr)}, the highest population was {name} ({int(count)})")
````
