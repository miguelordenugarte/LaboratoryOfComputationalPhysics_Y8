````python
import os
import inspect
import ast
````

````python
def save_functions(target):
    def functions(data):
        # We first try to see if the python file exists if not we create it.
        try:
            # When the file exists we read it to see the content.
            # r is for read only.
            with open(target, 'r') as f:
                content = f.read()
            if str(data) in content:
                 # if the function already exist in the file we do nothing with it.
                 return
            else:
                # if the function doesn't exist in the file we write it in the file.
                with open(target, 'a') as f:
                    f.write(str(data) + '\n')
        except FileNotFoundError:    
            with open(target, 'a') as f:
                f.write(str(data) + '\n')
    return functions
````

````python
save = save_functions('functions.py')
````

4\. Write the factorial function so that it a) does and b) does not use recursion.

````python
from functools import reduce
````

````python
def factorial(n):
    l = []
    x_iter = iter(range(1,n+1))
    for x in x_iter:
        l.append(x)
    f = reduce(lambda k, l: k*l, l)
    return f
````

````python
factorial(3)
````

````python
def fact(n):
    x = range(1, n+1)
    f = reduce(lambda k, l: k*l, x)
    return f
````

````python
fact(5)
````

5\. Use HOFs (zip in particular) to compute the weight of a circle, a disk and a sphere, assuming different radii and different densities:

```python
densities = {"Al":[0.5,1,2],"Fe":[3,4,5],"Pb": [15,20,30]}
radii = [1,2,3]
```

where the entries of the dictionary's values are the linear, superficial and volumetric densities of the materials respectively.

In particular define a list of three lambda functions using a comprehension that computes the circumference, the area and the volume for a given radius.


````python
import math
````

````python
densities = {'Al': [0.5, 1, 2], 'Fe': [3, 4, 5], 'Pb': [15, 20, 30]}
radii = [1, 2, 3]

circumference  = [2*math.pi*r for r in radii]
area = [math.pi*(r**2) for r in radii]
volume = [(4/3)*math.pi*(r**3) for r in radii]

for l, a, v in zip(circumference, area, volume):
    for key in densities:
        circle = l * densities[key][0]
        disk = a * densities[key][1]
        sphere = v * densities[key][2]
        print(f'The weight of the circle made of {key} with circunference {l} is: {circle}')
        print(f'The weight of the circle made of {key} with area {a} is: {disk}')
        print(f'The weight of the circle made of {key} with volume {v} is: {sphere}')
````

````python
# Parent class
class Dog:

    # Class attribute
    species = 'mammal'

    # Initializer / Instance attributes
    def __init__(self, name, age, hungry):
        self.name = name
        self.age = age
        self.hungry = hungry

    # instance method
    def description(self):
        return "{} is {} years old".format(self.name, self.age)

    # instance method
    def speak(self, sound):
        return "{} says {}".format(self.name, sound)
    
    # eat method
    def eat(self):
        self.hungry = False
        return self.hungry

# Child class (inherits from Dog class)
class RussellTerrier(Dog):
    def run(self, speed):
        return "{} runs {}".format(self.name, speed)

# Child class (inherits from Dog class)
class Bulldog(Dog):
    def run(self, speed):
        return "{} runs {}".format(self.name, speed)
````

````python
def dogs(name, age, hungry):
    d = []
    p = []
    j = 0
    if isinstance(name, (list, tuple)) and isinstance(age, (list, tuple)) and isinstance(hungry, (list, tuple)):
        for i in range(len(name)):
            d.append(Dog(name[i], age[i], hungry[i]))
            p.append(f'{d[i].name} is {d[i].age}.')
            if hungry[i] is True:
                j +=1
        if j == len(d):
            print(f'I have {len(d)}. {" ".join(p)} And they are all mammals, of course. My dogs are hungry')
        else:
            print(f'I have {len(d)}. {" ".join(p)} And they are all mammals, of course. My dogs are not hungry')
    else:
        for i in range(len(d)+1):
            d.append(Dog(name, age, hungry))
            p.append(f'{d[i].name} is {d[i].age}.')
            if hungry is True:
                j +=1
        if j == len(d):
            print(f'I have {len(d)}. {" ".join(p)} And they are all mammals, of course. My dogs are hungry')
        else:
            print(f'I have {len(d)}. {" ".join(p)} And they are all mammals, of course. My dogs are not hungry')
    return d
````
