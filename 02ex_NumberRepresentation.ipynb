```python
def convert_decimal_hexadecimal(n):
    hex_map = {'A': 10, 'B': 11, 'C': 12, 'D': 13, 'E': 14, 'F': 15}
    if n == 0:
        return "0x0"

    sign = '-' if n < 0 else ''
    n = abs(n)

    hexadecimal_digits = ''

    while n > 0:
        remainder = n % 16
        if remainder in hex_map.values():
            key = [k for k, v in hex_map.items() if v == remainder][0]
            hexadecimal_digits = str(key) + hexadecimal_digits
        else:
            hexadecimal_digits = str(remainder) + hexadecimal_digits
        n = n // 16
    return sign + '0x' + hexadecimal_digits
```

2\. Write a function that converts a 32 bit word into a single precision floating point (i.e. interprets the various bits as sign, mantissa and exponent)

```python
def single_precision(n):
    bias = 127
    exponent = int(n[1:9], 2)
    m = n [9:]
    matissa = 1 + sum(int(digit) * 2**-(i+1) for i, digit in enumerate(m))
    sign = -1 if int(n[0]) == 1 else +1

    return sign * matissa * (2**(exponent-bias))
```

(b) re-express the standard solution formula by multiplying top and bottom by $-b\mp\sqrt{b^2-4ac}$ and again find the solution for $a=0.001$, $b=1000$ and $c=0.001$. How does it compare with what previously obtained? Why?

(c) write a function that compute the roots of a quadratic equation accurately in all cases

```python
def quadratic_eq(a, b, c):
    square = ((b**2) - 4*a*c)**(1/2)
    x = [(-b + square)/(2*a), (-b - square)/(2*a)]
    return x
```

```python
# a:
quadratic_eq(0.001, 1000, 0.001)
```

```python
# b:
def quadratic_eq2(a, b, c):
    square = ((b**2) - 4*a*c)**(1/2)
    x = [2*c/(-b - square), 2*c/(-b + square)]
    return x
```

```python
quadratic_eq2(0.001, 1000, 0.001)
# This solution is more stable.
```

```python
# c:
import cmath

def quadratic_eq3(a, b, c):
    square = cmath.sqrt(b**2 - 4*a*c)
    if b > 0:
        x1 = (-b - square) / (2*a)
    else:
        x1 = (-b + square) / (2*a)
    x2 = c / (a * x1)
    return [x1, x2]
```

7\. Consider the integral of the semicircle of radius 1:
$$
I=\int_{-1}^{1} \sqrt(1-x^2) {\rm d}x
$$
which it's known to be $I=\frac{\pi}{2}=1.57079632679...$.
Alternatively we can use the Riemann definition of the integral:
$$
I=\lim_{N\to\infty} \sum_{k=1}^{N} h y_k 
$$

with $h=2/N$ the width of each of the $N$ slices the domain is divided into and where
$y_k$ is the value of the function at the $k-$th slice.

(a) Write a programe to compute the integral with $N=100$. How does the result compares to the true value?

(b) How much can $N$ be increased if the computation needs to be run in less than a second? What is the gain in running it for 1 minute? 


```python
import numpy as np

# a:

# def Riemann(N):
#     h = 2/N
#     i = -1
#     I = 0
#     for k in range(1, N + 1):
#         x = np.arange(i, i+h, 1)
#         sol = h * (1 - (x**2))**(0.5)
#         I = I + sol
#         i = i + h
#     return I

def Riemann(N):
    h = 2 / N
    I = 0
    for k in range(1, N + 1):
        x_k = -1 + k * h
        y_k = np.sqrt(1 - x_k**2)
        I += h * y_k
    return I

approx = Riemann(100)
print("Approximation:", approx)
print("True value (π/2):", np.pi / 2)
print("Error:", abs(approx - np.pi / 2))
```

```python
for N in range(700000, 10000000, 100000):
    start = time.perf_counter()
    Riemann(N)
    end = time.perf_counter()
    if end - start > 60:
        print("Max N under 1 minute:", N_previous)
        break
    N_previous = N
```

### (b) Discussion of Runtime and Accuracy

For  
$$ N \approx 7\times10^{5}, $$
the program takes about **1 second** to run. Increasing N further increases accuracy, but the improvement slows down because the **Riemann method converges as** O(1/N). This means the error decreases only proportionally to 1/N: doubling N roughly halves the error, but also doubles the computation time. For example, running the program for about **1 minute** allows  
$$ N \approx 9\times10^{6}, $$
improving the result by roughly two orders of magnitude. However, the overall accuracy does not change drastically due to the slow convergence rate of the Riemann sum.
