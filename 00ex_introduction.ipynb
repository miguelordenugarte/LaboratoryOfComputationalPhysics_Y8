1. The MickeyMouse problem

a) Write a program that prints the numbers from 1 to 100. But for multiples of 3 print `Mickey` instead of the corresponding number and for the multiples of 5 print `Mouse`. For numbers which are multiples of both three and five print `MickeyMouse`

b) Put the result in a tuple and substitute `Mickey` with `Donald` and `Mouse` with `Duck`

```python
lista = []
for i in range(1,101):
    if i % 3 == 0 and i % 5 == 0:
        a = "MickeyMouse"
        print(a)
        lista_2 = [a]
    else:
        if i % 3 == 0:
            b = "Mickey"
            print(b)
            lista_2 = [b]
        elif i % 5 == 0:
            c = "Mouse"
            print(c)
            lista_2 = [c]
        else:
            print(i)
            lista_2 = [i]
    lista += lista_2

lista = ["Donald" if x == "Mickey" else "Duck" if x == "Mouse" else x for x in lista] #We substitute the words in the list because tuples are inmutable.
tupla = tuple(lista)
print(tupla)
```

2\. The swap function

Write a function that swap the values of two input variables x and y (whatever the type). Try to do that also without a temporary variable

```python
def swap_numbers(x = None, y = None):
    if x is None:
        x = input("Value of x")
    if y is None:
        y = input("Value of y")
    x, y = y, x
    return x, y
```

```python
# First we try with a global variable.
x = 5
y = 2
swap_numbers(x,y)
```

```python
# We try now with a temporary variable we define inside the function with an input.
swap_numbers()
```

3\. Computing the distance

Write a function that calculates and returns the euclidean distance between two points *u* and *v*, where *u* and *v* are both 2-tuples *(x,y)*. For example, if *u=(3,0)* and *v=(0,4)*, the function should return 5

```python
def euclidean_distance(u, v):
    diference = ()
    for i in range(0, len(u)):
        dif = (u[i] - v[i],)
        diference = diference + dif
    distance = ((diference[0])**2 + (diference[1])**2)**(1/2)
    print(f"The euclidean distance between u and v is: {distance}")
```

```python
euclidean_distance((3,0),(0,4))
```

4\. Counting letters

Write a program to calculate the number of times each character occurs in a given string *s*. Ignore differneces in capitalization

```python
s="Write a program that prints the numbers from 1 to 100. \
But for multiples of three print Mickey instead of the number and for the multiples of five print Mouse. \
For numbers which are multiples of both three and five print MickeyMouse"
```

```python
def count_characters(c):
    c = c.lower().replace(" ", "")
    unique = set(c) # We use set in order to get the unique characters.
    unique_characters = list(unique)
    initial = 0
    count = dict.fromkeys(unique_characters, initial) # We create a dictionary with the unique characters and we set their initial value to 0.
    for i in range(0, len(c)):
        if c[i] in unique_characters:
            count[c[i]] += 1
    return count
```

```python
count_characters(s)
```

6\. Combination of functions

Write two functions - one that returns the square of a number, and one that returns the cube. Now write a third function that returns the number raised to the 6th power using the two previous functions.

```python
def square(n):
    return n**2

def cube(n):
    return n**3
```

```python
def sixth_power(n):
    return cube(square(n))
```

7\. Cubes

Create a list of the cubes of x for x in *[0, 10]* using:

a) a for loop

b) a list comprehension

```python
n = list(range(11))

# loop case:

def loop_cube(n):
    cubes = []
    for i in range(0, len(n)):
        cub = n[i]**3
        cubes.append(cub)
    return cubes
```

```python
loop_cube(n)
```

```python
def list_cube(n):
    cube = [n[i]**3 for i in range(0, len(n))]
    return cube
```

```python
list_cube(n)
```

8\. Nested list comprehension

A Pythagorean triple is an integer solution to the Pythagorean theorem $a^2+b^2=c^2$. The first Pythagorean triple is (3,4,5). Find and put in a tuple all unique Pythagorean triples for the positive integers a, b and c less than 100.

```python
n = list(range(101))
def pythagorean_triple(n):
    tuple_num = ()
    for a in range(1, len(n)):
        for b in range(1, len(n)):
            for c in range(1, len(n)):
                if a**2 + b**2 == c**2:
                    m = [a,b,c]
                    tupla = (m,)
                    tuple_num = tuple_num + tupla
    return tuple_num
```

```python
pythagorean_triple(n)
```

9\. Normalization

Write a function that takes a tuple of numbers and returns it with the entries normalized to one

```python
def normalized(l):
    l = list(l)
    norm = [(x - min(x)) / (max(x) - min(x)) for x in l]
    return norm
```

```python
n = (1,5,6,7)
normalized(n)
```
